```python
import numpy as np
import lazyplot
from PIL import Image
```

```python
x = np.random.rand(3, 30)
lazyplot.lazy_plot(x)
```

```python
a1 = lazyplot.LazyAxes(np.random.rand(30, 30), title="a1", plot_type="imshow")
a2 = lazyplot.LazyAxes(np.random.rand(3, 30), plot_type="plot")

data = lazyplot.custom_plot([a1, a2], figure_config={"figsize": (5,10)})
data.show()
```
